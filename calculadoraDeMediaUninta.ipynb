# Calculadora de média por disciplina (Uninta).

### **Legenda**

**AP1** = Prova parcial 1.  
**AP1A** = Autoavaliação da prova 1.  
**AP2** = Prova parcial 2.  
**AP2A** = Autoavaliação da prova 2.  
**AP3** = Prova parcial 3.  
**AP3D** = Prova parcial 3 escrita.  

### **AVISO IMPORTANTE!!!**

Ao invés de usar vírgula, use o ponto final em números quebrados.

```python
ap1 = float(input("Digite a nota da AP1: "))
ap1a = float(input("Digite a nota da AP1A: "))

ap2 = float(input("Digite a nota da AP2: "))
ap2a = float(input("Digite a nota da AP2A: "))

ap3 = float(input("Digite a nota da AP3: "))
ap3d = float(input("Digite a nota da AP3D: "))

media_final = (
    (ap1 * 0.1) +
    (ap1a * 0.1) +
    (ap2 * 0.1) +
    (ap2a * 0.1) +
    (ap3 * 0.4) +
    (ap3d * 0.2)
)

print(f"Média final: {media_final:.2f}")
```
